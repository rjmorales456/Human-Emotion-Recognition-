```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPool2D,Flatten,Dense
from tensorflow.keras.preprocessing import image
from tensorflow import keras
import os
import cv2 as cv
import random
```

```python
#Load Haar Cascade
cascade = cv.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
model = tf.keras.models.load_model('trial.h5')
```

Loading Model

```python
# detect video feed 
video = cv.VideoCapture(0)
while True:
    ret, frame = video.read()
    
    # grayscale
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    print(gray.shape)
    
    x = 100
    y = 150
    face_recognition = cascade.detectMultiScale(gray,1.1,7)
    
    for (x,y,width,height) in face_recognition:
        face_gray = gray[y:y+height,x:x+width]
        face_color = frame[y:y+height,x:x+width]
        
        cv.rectangle(frame,(x,y),(x+width,y+height),(0,255,0),thickness=2)
        
        face = cascade.detectMultiScale(face_gray)
        
        # if face is detected
        
        if len(face) == 0:
            print("Face is not Detected")
            
            
        # finding the emotion
        
        vid = cv.resize(face_color,[128,128])
        vid = np.expand_dims(vid, axis=0)
            
        vid = vid / 255.0
            
            
        pred = model.predict(vid)      
        print(pred[0])
            
        #making a prediction
        emotion_index =  np.argmax(pred)
        emotion_labels = ["Angry", "Fear", "Happy", "Neutral", "Sad", "Surprised"]
        emotion = emotion_labels[emotion_index]
        
        #video
        cv.putText(frame,emotion,(x, y - 10),cv.FONT_HERSHEY_COMPLEX,1,(0,255,0),thickness=2)
        cv.rectangle(frame,(x,y),(x+width,y+height),(0,255,0),thickness=2)
        print(emotion)
            
    cv.imshow("Video Frame",frame)
    
    if cv.waitKey(1)==ord('q'):
        break   
    
video.release()
cv.destroyAllWindows()
            
            
```
